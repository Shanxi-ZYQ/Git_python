```python
import requests
from requests.structures import CaseInsensitiveDict
from IPython.display import Image , display
from IPython.core.display import HTML
from lxml import etree

# next_page_url = "https://book.douban.com/tag/%E7%A5%9E%E7%BB%8F%E7%BD%91%E7%BB%9C"
page = 1
book_info = []

while 1:
    
    next_page_url = f"https://book.douban.com/tag/%E7%A5%9E%E7%BB%8F%E7%BD%91%E7%BB%9C?start={(page-1)*20}&type=T"
    headers = CaseInsensitiveDict()
    headers["User-Agent"] = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.60 Safari/537.36"

    r = requests.get(next_page_url, headers=headers)
    content = r.text
    sel = etree.HTML(content)

    for block in sel.xpath("//li[@class='subject-item']"):
        title = ''
        elem_title = block.xpath('.//h2/a')
        if elem_title:
            title = ''.join(elem_title[0].itertext()).replace(' ','').replace('\n','')
        print(title)
        
        price = -1
        elem_price = block.xpath(".//span[@class='buy-info']/a/text()")
        if elem_price:
            s = elem_price[0].strip()
            price = float(s[s.find('版')+1:s.find('元')])
        print(price)

        cover = ''
        elem_cover = block.xpath(".//img/@src")
        if elem_cover:
            cover = elem_cover[0]
        # display(HTML(f"<img src={cover}>"))

        book_info.append([title, price, cover])
    # elem_next_page = sel.xpath("//span[@class='next']/a/@href")
    # if elem_next_page:
    #     next_page_url = "https://book.douban.com"+elem_next_page[0]
    # else:
    #     break
    page += 1
    if page > 4:
        break
```
